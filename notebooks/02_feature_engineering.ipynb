"""
Feature Engineering for the Home Credit Default Risk dataset.

This notebook creates applicant-level features from the main application data
and the secondary Home Credit tables. It then cleans the combined dataset and
saves processed train/test files for model training.
"""

```python
from pathlib import Path
import sys

import pandas as pd
```

```python
PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.config import (
    ID_COLUMN,
    TARGET_COLUMN,
    PROCESSED_DATA_DIR,
    PROCESSED_TRAIN_FILE,
    PROCESSED_TEST_FILE,
)

from src.data_loader import load_all_raw_data
from src.feature_engineering import build_full_feature_dataset
from src.data_cleaning import clean_application_data
```

```python
pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 100)
```

```python
data = load_all_raw_data()

print("Raw data loaded successfully.")
print(f"application_train shape: {data['application_train'].shape}")
print(f"application_test shape: {data['application_test'].shape}")
```

```python
train_features, test_features = build_full_feature_dataset(data)

print("Feature engineering completed.")
print(f"Train features shape before cleaning: {train_features.shape}")
print(f"Test features shape before cleaning: {test_features.shape}")
```

```python
train_features.head()
```

```python
test_features.head()
```

```python
processed_train, processed_test, y, cleaning_metadata = clean_application_data(
    train_df=train_features,
    test_df=test_features,
    target_column=TARGET_COLUMN,
    id_column=ID_COLUMN,
)

print("Cleaning completed.")
print(f"Processed train shape: {processed_train.shape}")
print(f"Processed test shape: {processed_test.shape}")
print(f"Target shape: {y.shape}")
```

```python
processed_train[TARGET_COLUMN] = y.values

print("Target added back to processed training dataset.")
processed_train.head()
```

```python
print(f"Number of columns removed due to high missingness: {len(cleaning_metadata['high_missing_columns'])}")
print(f"Number of numeric columns before encoding: {len(cleaning_metadata['numeric_columns'])}")
print(f"Number of categorical columns before encoding: {len(cleaning_metadata['categorical_columns'])}")
print(f"Final number of feature columns: {len(cleaning_metadata['feature_columns'])}")
```

```python
cleaning_metadata["high_missing_columns"][:30]
```

```python
missing_after_cleaning = processed_train.isna().mean().sort_values(ascending=False)

missing_after_cleaning.head(20)
```

```python
processed_train[TARGET_COLUMN].value_counts(normalize=True)
```

```python
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)

processed_train_path = PROCESSED_DATA_DIR / PROCESSED_TRAIN_FILE
processed_test_path = PROCESSED_DATA_DIR / PROCESSED_TEST_FILE

processed_train.to_csv(processed_train_path, index=False)
processed_test.to_csv(processed_test_path, index=False)

print(f"Processed train saved to: {processed_train_path}")
print(f"Processed test saved to: {processed_test_path}")
```

```python
metadata_summary = pd.DataFrame(
    {
        "metric": [
            "processed_train_rows",
            "processed_train_columns",
            "processed_test_rows",
            "processed_test_columns",
            "high_missing_columns_removed",
            "numeric_columns_before_encoding",
            "categorical_columns_before_encoding",
            "final_feature_columns",
        ],
        "value": [
            processed_train.shape[0],
            processed_train.shape[1],
            processed_test.shape[0],
            processed_test.shape[1],
            len(cleaning_metadata["high_missing_columns"]),
            len(cleaning_metadata["numeric_columns"]),
            len(cleaning_metadata["categorical_columns"]),
            len(cleaning_metadata["feature_columns"]),
        ],
    }
)

metadata_summary
```

```python
metadata_summary.to_csv(
    PROCESSED_DATA_DIR / "feature_engineering_summary.csv",
    index=False,
)

pd.Series(cleaning_metadata["high_missing_columns"]).to_csv(
    PROCESSED_DATA_DIR / "removed_high_missing_columns.csv",
    index=False,
)

print("Feature engineering summary files saved successfully.")
```
