"""
Exploratory Data Analysis for the Home Credit Default Risk dataset.

This notebook validates the raw dataset, inspects table sizes, checks the target
distribution, reviews missing values, and confirms how the secondary tables link
to the main application data.
"""

```python
from pathlib import Path
import sys

import pandas as pd
import matplotlib.pyplot as plt
```

```python
PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.config import ID_COLUMN, TARGET_COLUMN, REPORTS_DIR
from src.data_loader import load_all_raw_data, get_dataset_shapes, print_dataset_shapes
```

```python
pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 100)
```

```python
data = load_all_raw_data()
print_dataset_shapes(data)
```

```python
table_shapes = get_dataset_shapes(data)
table_shapes
```

```python
application_train = data["application_train"]
application_test = data["application_test"]

application_train.head()
```

```python
application_train.info()
```

```python
target_distribution = (
    application_train[TARGET_COLUMN]
    .value_counts()
    .rename_axis("target")
    .reset_index(name="count")
)

target_distribution["percentage"] = (
    target_distribution["count"] / target_distribution["count"].sum()
).round(4)

target_distribution
```

```python
plt.figure(figsize=(6, 4))
plt.bar(
    target_distribution["target"].astype(str),
    target_distribution["percentage"],
)
plt.title("Target Distribution")
plt.xlabel("TARGET")
plt.ylabel("Percentage")
plt.show()
```

```python
missing_summary = pd.DataFrame(
    {
        "column": application_train.columns,
        "missing_count": application_train.isna().sum().values,
        "missing_percentage": application_train.isna().mean().values,
        "dtype": application_train.dtypes.astype(str).values,
    }
)

missing_summary = missing_summary.sort_values(
    by="missing_percentage",
    ascending=False,
).reset_index(drop=True)

missing_summary.head(30)
```

```python
numeric_columns = application_train.select_dtypes(include=["number"]).columns.tolist()

numeric_summary = application_train[numeric_columns].describe().T.reset_index()
numeric_summary = numeric_summary.rename(columns={"index": "column"})

numeric_summary.head(30)
```

```python
secondary_table_coverage = []

main_ids = set(application_train[ID_COLUMN].dropna().unique())

for table_name, df in data.items():
    if table_name in ["application_train", "application_test"]:
        continue

    if ID_COLUMN not in df.columns:
        continue

    table_ids = set(df[ID_COLUMN].dropna().unique())
    overlapping_ids = main_ids.intersection(table_ids)

    secondary_table_coverage.append(
        {
            "table_name": table_name,
            "rows": len(df),
            "unique_applicants": len(table_ids),
            "overlapping_train_applicants": len(overlapping_ids),
            "coverage_of_table_applicants": (
                len(overlapping_ids) / len(table_ids)
                if len(table_ids) > 0
                else 0
            ),
        }
    )

coverage_df = pd.DataFrame(secondary_table_coverage)
coverage_df
```

```python
for table_name, df in data.items():
    print(f"\n--- {table_name} ---")
    display(df.head())
```

```python
REPORTS_DIR.mkdir(parents=True, exist_ok=True)

table_shapes.to_csv(REPORTS_DIR / "eda_table_shapes.csv", index=False)
target_distribution.to_csv(REPORTS_DIR / "eda_target_distribution.csv", index=False)
missing_summary.to_csv(REPORTS_DIR / "eda_missing_summary_application_train.csv", index=False)
coverage_df.to_csv(REPORTS_DIR / "eda_secondary_table_coverage.csv", index=False)
```
